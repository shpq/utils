```python
import os
import shutil
from tqdm.notebook import tqdm
from PIL import Image
```

```python
path = "storages/smiles/smile_gen_resized"
names = [x for x in os.listdir(path) if x.endswith(".jpg")]
```

```python
rand_path = "storages/smiles/random"
os.makedirs(rand_path, exist_ok=True)
```

```python
smile_path = "storages/smiles/smile"
os.makedirs(smile_path, exist_ok=True)
```

```python
no_smile_path = "storages/smiles/no_smile"
os.makedirs(no_smile_path, exist_ok=True)
```

```python
for rand_n in tqdm([x for x in names if "orig" in x]):
    shutil.copy2(
        os.path.join(path, rand_n),
        os.path.join(rand_path, rand_n),
    )
```

```python
for smile in tqdm([x for x in names if "target2" in x]):
    shutil.copy2(
        os.path.join(path, smile),
        os.path.join(smile_path, smile),
    )
```

```python
for el in tqdm([x for x in os.listdir(rand_path) if x.endswith(".jpg")]):
    if "-" in el:
        break
```

```python
for el in tqdm([x for x in os.listdir(rand_path) if x.endswith(".jpg")]):
    if "-" in el:
        shutil.copy2(
            os.path.join(rand_path, el),
            os.path.join(no_smile_path, el),
        )
```

```python
seeds = [x.split("_")[3] for x in os.listdir(no_smile_path)]
```

```python
smile_paired_path = "storages/smiles/smile_paired"
os.makedirs(smile_paired_path, exist_ok=True)
```
